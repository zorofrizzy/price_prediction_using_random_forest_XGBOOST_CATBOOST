```python
#Checking bindings from the JSON file and then substituting for categorical data
with open(r'F:\Kabir\CSUN\Comp542\label_encoding_mappings_team_4.json') as f:
    data = json.load(f)

def key_search(col_value): 
    temp_list = []
    for key in data.keys():
        if key in col_value:
            for k, v in data[key].items():
                temp_list.append([k, v])
            return temp_list

            
```

```python
def bar_chart(col_value, variable_name):
    fig, ax = plt.subplots(figsize=(10, 5))
    plt.bar(df[col_value], df['price'], width = 0.1, edgecolor = "black")
    plt.title("Price X {} Bar Plot".format(variable_name))
    plt.xlabel(variable_name)
    plt.ylabel("Price (USD)")
    
    
```

```python
def scatter_plot(col_value, variable_name):
    plt.scatter(x = df[col_value], y = df['price'])
    plt.title("Price X {} Scatter Plot".format(variable_name))
    plt.xlabel(variable_name)
    plt.ylabel("Price (USD)")
    
```

```python
scatter_plot('humidity', "Humidity")
plt.show()

scatter_plot('visibility', "Visibility")
plt.show()

scatter_plot('windBearing', "Wind Bearing")
plt.show()

#Bar Graphs

scatter_plot('moonPhase', "Moon Phase")
plt.show()

scatter_plot('precipIntensityMax', "Precipitation Intensity Maxima")
plt.show()

scatter_plot('distance', "Distance")
plt.show()

```
