```python
import pandas as pd
```

```python
df = pd.read_csv(r'F:\Kabir\CSUN\Comp542\preprocessed_csv_team_4.csv')
```

```python
# Binning the data into 9 bins of bin size = 5

def binner(bin_df):
    x = 0
    if bin_df <= 5:
        x = 1
    elif bin_df > 5 and bin_df <= 10:
        x = 2
    elif bin_df > 10 and bin_df <= 15:
        x = 3
    elif bin_df > 5 and bin_df <= 20:
        x = 4
    elif bin_df > 5 and bin_df <= 25:
        x = 5
    elif bin_df > 5 and bin_df <= 30:
        x = 6
    elif bin_df > 5 and bin_df <= 35:
        x = 7
    elif bin_df > 5 and bin_df <= 40:
        x = 8
    elif bin_df > 5 and bin_df <= 45:
        x = 9
    return x

df['price_bin'] = df['price'].apply(binner)

    
```

```python
x = list(df['price_bin'].value_counts().items())
```

```python
x
```

```python
bin_dict = {}
list1 = []
list2 = []

for entity in x:
    list1.append(entity[0])
    list2.append(entity[1])
    
total_entities = sum(list2)

#Storing ratio in bin dict

for i in range(len(list1)):
    bin_dict[list1[i]] = int(float(list2[i]/total_entities)*5000) #5000 here since total we will work with 5K entries
print(bin_dict)
```

```python
# Selecting DFs based on the condition that the price_bin matches and we pick up the required number of rows for each
# price bin.

new_df = pd.DataFrame()
new_list = []
for i in list1:
    new_list.append(df[df['price_bin'] == i].head(bin_dict[i]))

new_df = new_df._append(new_list, ignore_index = True)
new_df
```

```python
new_df.describe().T
```

```python
# Checking if the minimum and maximum values from our original DF are copied or not. 
# If not, we copy them.

min_price = df.price.min()
max_price = df.price.max()
print("Original DF Min Price : {}\nOriginal DF Max Price : {}".format(min_price, max_price))

new_min = new_df.price.min()
new_max = new_df.price.max()

if new_min > min_price:
    #Copy one instance
    new_df = new_df._append(df[df['price'] == min_price].head(1), ignore_index = True)
    
if new_max < max_price:
    #Copy one instance
    new_df = new_df._append(df[df['price'] == max_price].head(1), ignore_index = True)

new_min = new_df.price.min()
new_max = new_df.price.max()
print("New DF Min Price : {}\nNew DF Max Price : {}".format(new_min, new_max))
```

```python
# Deleting the column for Price_bin 
if 'price_bin' in new_df.columns:
    
    new_df = new_df.drop('price_bin', axis = 1) #Axis 0 for rows and 1 for columns.

new_df
```
